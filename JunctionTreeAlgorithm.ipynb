## Junction Tree Algorithm

```python
# import essential package
# THIS CODE NEED TO RUN IN PYTHON 3.X
import numpy as np
```

```python
# current potential of cliques
potential_12 = np.array([[0.1, 0.7], [0.8, 0.3]])
potential_23 = np.array([[0.5, 0.1], [0.1, 0.5]])
potential_34 = np.array([[0.1, 0.5], [0.5, 0.1]])
potential_45 = np.array([[0.9, 0.3], [0.1, 0.3]])

ini_seperator = np.array([1., 1.])
# randomlt select one node as root
# root:clique x4x5
# initialization
cliques = {
    1:potential_12,
    2:potential_23,
    3:potential_34,
    4:potential_45
}

seperator = {
    2:ini_seperator,
    3:ini_seperator,
    4:ini_seperator
}
# collect and distribute 
print('initialize:--------')
print(cliques)

# collect stage
# update seperator and clique potential
def collect(previous_clique, current_clique):
    old_seperator = np.array([1., 1.])
    old_clique = current_clique
    new_seperator = np.sum(previous_clique, axis=0)
    new_clique = np.row_stack(((new_seperator / old_seperator)[0] * old_clique[0, :],
                               (new_seperator / old_seperator)[1] * old_clique[1, :]))
    print(new_seperator)
    print(new_clique,"\n")
    return new_clique, new_seperator

print('collect:-------------')
for index in range(1, len(cliques)):
    print(index, ':')
    cliques[index + 1], seperator[index + 1] = collect(cliques[index], cliques[index + 1])
    

    
    # distribute stage
def distribute(old_seperator, previous_clique, current_clique):
#     old_seperator = seperator[4]
#     previous_clique = cliques[4]
#     current_clique = cliques[3]
    old_clique = current_clique
    new_seperator = np.sum(previous_clique, axis=1)
    new_clique = np.column_stack(((new_seperator / old_seperator)[0] * old_clique[:, 0],
                               (new_seperator / old_seperator)[1] * old_clique[:, 1]))

    print(new_seperator)
    print(new_clique,"\n")
    return new_clique, new_seperator

print('distribute:-------------')
for index in range(len(cliques), 1, -1):
    print(index, ':')
    cliques[index - 1], seperator[index] = distribute(seperator[index], cliques[index], cliques[index - 1])


print('normalize:--------------')
for index in range(len(cliques)):
    cliques[index+1] = cliques[index + 1] / np.sum(cliques[index + 1])
    if index != 3:
        seperator[index+2] = np.sum(cliques[index+1], axis=0)

prob_1 = np.sum(cliques[1], axis=1)
prob_5 = np.sum(cliques[4], axis=0)


for i in range(1, len(cliques)+1):
    print('cliquex{}{}:\n{}'.format(i, i+1, cliques[i]))

print('')
print('p(x{}):{}'.format(1, prob_1))
for i in range(2, len(seperator)+2):
    print('p(x{}):{}'.format(i, seperator[i]))
print('p(x{}):{}'.format(5, prob_5))
```
